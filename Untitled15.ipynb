```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
```

```python
input_csv_name = "twitter_training.csv"  
output_csv_name = "sentiment_results_textblob.csv"
```

```python
df = pd.read_csv(input_csv_name)
```

```python
sia = SentimentIntensityAnalyzer()

```

```python
def analyze_sentiment(text):
    text = str(text)
    score = sia.polarity_scores(text)
    numeric = score['pos'] - score['neg']  # simple numeric score
    # Determine sentiment label based on compound score
    if score['compound'] >= 0.05:
        label = 'positive'
    elif score['compound'] <= -0.05:
        label = 'negative'
    else:
        label = 'neutral'
    return numeric, label
```

```python
df[['sentiment_score','sentiment_label']] = df['review'].apply(analyze_sentiment).apply(pd.Series)


df.to_csv(output_csv_name, index=False)
print("Sentiment analysis complete. Results saved to:", output_csv_name)

print(df.head())
```

```python
summary = df['sentiment_label'].value_counts()
print("Sentiment Summary:\n", summary)
```

```python
summary.to_csv("sentiment_summary.csv", header=True)
print("Summary table saved as sentiment_summary.csv")
```

```python
plt.figure(figsize=(6,4))
summary.plot(kind='bar', color=['red', 'grey', 'green'])
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig("sentiment_distribution.png")  # save chart
plt.show()
```
